# SQL Questions

```python
#Find orders shipped to the city = "montes claros"
query='''SELECT customer_unique_id,customer_zip_code_prefix,customer_city 
FROM olist_customers_dataset
WHERE customer_city="montes claros";'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Count orders per order status
query='''SELECT order_status,COUNT(order_id) AS NUMBER_OF_ORDERS
FROM olist_orders_dataset
GROUP BY order_status
ORDER BY NUMBER_OF_ORDERS DESC;'''
cursor.execute(query)
cursor.fetchall()
```

```python
#List products and their categories for all order items
query='''SELECT items.order_id,items.product_id,product.product_category_name
FROM olist_order_items_dataset items INNER JOIN olist_products_dataset product
ON items.product_id=product.product_id;'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Find all sellers and their total sales amount
query='''SELECT seller.seller_id,SUM(payment.payment_value) AS TOTAL_INCOME,seller.seller_city
FROM olist_order_items_dataset items RIGHT JOIN olist_sellers_dataset seller
ON items.seller_id=seller.seller_id 
LEFT JOIN olist_order_payments_dataset payment
ON items.order_id=payment.order_id
GROUP BY seller.seller_id,seller.seller_city
ORDER BY TOTAL_INCOME DESC;'''
cursor.execute(query)
cursor.fetchall()
```

```python
#List all orders and their reviews
query='''SELECT review.order_id,orders.order_status,review.review_score
FROM olist_orders_dataset AS orders Right JOIN olist_order_reviews_dataset AS review
ON orders.order_id=review.order_id;'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Find customers who spent more than the average order value
query='''SELECT orders.customer_id
FROM olist_orders_dataset orders INNER JOIN olist_order_payments_dataset payment
ON orders.order_id=payment.order_id
GROUP BY orders.customer_id
HAVING SUM(payment.payment_value)>
(
SELECT AVG(payment_value) FROM olist_order_payments_dataset);'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Get the most expensive product sold
query='''SELECT order_id
FROM olist_order_items_dataset 
WHERE price=(
SELECT MAX(price) FROM olist_order_items_dataset);'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Calculate average review score per product category
query='''SELECT product.product_category_name,AVG(review.review_score) AS AVERAGE_SCORE
FROM olist_products_dataset AS product INNER JOIN olist_order_items_dataset AS orders
ON orders.product_id=product.product_id
INNER JOIN olist_order_reviews_dataset AS review
ON orders.order_id=review.order_id
GROUP BY product.product_category_name;'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Create a view of top 10 sellers by total revenue
query='''SELECT order_id,SUM(price) AS REVENUE
FROM olist_order_items_dataset 
GROUP BY order_id
ORDER BY REVENUE;'''
cursor.execute(query)
cursor.fetchall()
```

```python
#Create indexes to speed up joins and filtering
query='''CREATE INDEX idx_orders_customer ON olist_orders_dataset(customer_id(50));
CREATE INDEX idx_order_items_product ON olist_order_items_dataset(product_id);
CREATE INDEX idx_payments_order ON olist_order_payments_dataset(order_id);
CREATE INDEX idx_reviews_order ON olist_order_reviews_dataset(order_id);'''
cursor.execute(query)
cursor.fetchall()
```
